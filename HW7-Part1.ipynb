### Part Two

Scrape the content of https://www.lemonde.fr/Links to an external site. and save it as a CSV.

We want: titles, subhead, article URL, whether it's premium or not, byline, article type, image URL.

Bonus, if you want to get fancy:

Make the CSV file auto-updating. Use this tutorial (videoLinks to an external site., textLinks to an external site.) but just ignore the visualization/datawrapper aspect


```python
import requests
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get("https://www.lemonde.fr")
doc = BeautifulSoup(response.content)
```

```python
#Titles, subhead, article URL, whether it's premium or not, byline, article type, image URL.
# Find all of the divs with the class of news
# items = doc.find_all('div', attrs={'class': 'news'})
# items = doc.find_all('div', class_='news')
# items = doc.find_all(class_='news')
# .news means "the thing with the class of news"
items = doc.select(".article.article--featured.old__article-square")
len(items)    
```

```python
print(items)
```

```python
articles = []
for item in items:
    title = item.select_one('.article__title.article__title--inline', None).text
    subheader = item.select_one('.article__desc.article__subheader', None)
    atype = item.select_one('.article__type', None).text
    byline = item.select_one('.article__byline', None).text
    url = item.select_one('a')['href']
    img = item.select_one('img').get('data-src', None)
    article = {
        'url': url,
        'title': title,
        'subtitle': subheader,
        'type': atype,
        'byline': byline,
         'img_path': img
    }
    articles.append(article)
len(articles)
```

```python
import pandas as pd

df = pd.DataFrame(articles)
df.head()
```
