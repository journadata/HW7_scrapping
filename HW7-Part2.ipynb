```python
from bs4 import BeautifulSoup
import requests

response = requests.get("https://www.lemonde.fr")
doc = BeautifulSoup(response.content)
```

```python
doc.find('h1').text
```

```python

#Find all the divs with the class of article main

#items = doc.find('div', class_='article--main')
#If I want one thing I will use doc.select_one(".news")
items = doc.select(".article--main")

rows_main = []
#Loop through them and print their text
for item in items:
    print("----")
    #Find me the thing that has the class of "article__title-label" and print it out the text
    title = item.find(class_='article__title-label').text.strip()
    #Find me the thing that has the class of "article__subheader" and print it out the text
    print("----")
    subtitle = item.find(class_='article__subheader').text.strip()
    #Find me the thing that has the class of "article__subheader" and print it out the text
    print("-------")
    #Find me the link and get the URL it points to
    url = item.find('a')['href']
    print("-------")
    article_type = item.find('.article__type', None)
    print("-------")
    byline = item.find('.article__byline', None)
    print("-------")
    #Find me the thing with the tag name 'img' and print it out
    image = item.select_one('img').get('data-src', None)
      # We start saying each row has no data
    row = {
        'title': title, 
        'subtitle': subtitle, 
        'url': url, 
        'article_type': article_type,
        'byline': byline,
        'image': image        
    }

    rows_main.append(row)
```

```python
#Fins all the divs with the class of article main

rivers = doc.find_all('div', attrs={'class': 'old__article-river'})

#Loop through them and print their text
for river in rivers:
    print("1.----")
    print(river.find(class_='article__title').text.strip())
     #Find me the thing that has the class of "article__subheader" and print it out the text
    print("2.----")
    print(river.find(class_='article__subheader'), None)
    #Find me the thing that has the class of "article__subheader" and print it out the text
    print("3.-------")
    print(river.find('a')['href'])
    print("4.-------")
    print(river.find('sr-only', None))
    print("5.-------")
    print(river.find(class_='article__author').text.strip() if river.find(class_='article__author') else None)
    print("6.-------")
    #I had problems when trying to find the image, so I look for a solution in Google.
    #I found this, and I adapted it : https://stackoverflow.com/questions/70134620/get-image-data-src-with-beautiful-soup-when-there-is-no-image-extensiona
    print(river.select_one('img')['data-src'] if river.select_one('img') else None)
   
```

```python
rows_rivers = []

for river in rivers:
    print("----")
    title = river.find(class_='article__title').text.strip()
     #Find me the thing that has the class of "article__subheader" and print it out the text
    print("----")
    subtitle = river.find(class_='article__subheader'), None
    #Find me the thing that has the class of "article__subheader" and print it out the text
    print("-------")
    url = river.find('a')['href']
    print("-------")
    article_type = river.find('sr-only', None)
    print("-------")
    byline = river.find('article__author-name', None)
    print("-------")
    #I had problems when trying to find the image, so I look for a solution in Google.
    #I found this, and I adapted it : https://stackoverflow.com/questions/70134620/get-image-data-src-with-beautiful-soup-when-there-is-no-image-extensiona
    image = river.select_one('img')['data-src'] if river.select_one('img') else None

    row = {
        'title': title, 
        'subtitle': subtitle, 
        'url': url, 
        'article_type': article_type,
        'byline': byline,
        'image': image        
    }

    rows_rivers.append(row)

```
